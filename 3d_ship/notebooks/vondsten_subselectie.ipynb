```python
import pandas as pd
```

```python
# Panda settings for showing data (this is foremost done to more easily explore the data while processing it)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
path_to_folder = "C:\\path\\to\\folder\\"
path_to_excel = path_to_folder+"Overzicht alle data.xlsx"
df = pd.read_excel(path_to_excel)
```

```python
df.head()
```

```python
df_sub = df[df["Spooraard"] == "bun dl2"].copy()
```

```python
df_sub["VondstNr"] = df_sub["VondstNr"].astype(int)
df_sub["Vaknr"] = df_sub["Vaknr"].astype(int)
```

```python
df_sub.to_csv(path_to_folder+"sub_bun2.csv")
```

```python
df_sub.head(300)
```

```python
pivot_tbl = df_sub[['VondstNr', 'Soort' , 'AANTAL']] 
```

```python
pivotted = pivot_tbl.pivot_table(
    index="VondstNr",
    columns="Soort",
    values="AANTAL",
    aggfunc="sum",
    fill_value=0
)
```

```python
pivotted = pivotted.rename(columns={
    "Baars/pos": "Baars_pos",
    "Elft/fint": "Elft_fint",
    "Kleine zeenaald/grote zeenaald": "Kleine_zeenaald_grote_zeenaald",
    "Driedoornige stekelbaars":"Driedoornige_stekelbaars"
})
```
