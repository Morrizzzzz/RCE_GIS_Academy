```python
import pyogrio
```

```python
brp_path = 'data//'
brp_name = 'brpgewaspercelen_definitief_2025'
brp_updated = brp_path+brp_name+"_updated.gpkg"
```

```python
# the geopackage input file contains a field id which is not kept given its name, there uni_id is created
brp_org = pyogrio.read_dataframe(
    brp_path+brp_name+'.gpkg',
    layer="brp_gewas",
    fid_as_index=True
)

brp_org = brp_org.reset_index(names="uni_id")
brp_updated = brp_path+brp_name+"_updated.gpkg"

brp_org.to_file(
    brp_updated,
    layer="brp_gewas",
    driver="GPKG"
)
```
